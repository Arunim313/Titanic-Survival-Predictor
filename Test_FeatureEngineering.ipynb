```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv("test.csv")
```

```python
test.head()
```

```python
test.describe()
```

```python
sex =pd.get_dummies(test['Sex'],drop_first=True)
```

```python
embark =pd.get_dummies(test['Embarked'],drop_first=True)
```

```python
def ageFilling(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
test['Age']=test[['Age','Pclass']].apply(ageFilling,axis=1)
```

```python
test.drop(['Sex','Embarked','Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
test.drop('PassengerId',axis=1)
```

```python
test=pd.concat([test,sex,embark],axis=1)
```
